```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/"
response = requests.get(url,"html.parser")
soup = bs(response.text)
print(soup.prettify())
```

```python
info = soup.find("div",class_ = "articleContentBody")
info1 = info.find_all("div",class_ = "row countdown-item")
rank = []
title = []
year = []
score = []
director = []
cast = []
critics = []
for i in info1:
    i1 = i.text.strip().split("\n")
    rank.append(int(i1[0].split("#")[1]))
    title.append(i1[10].split(" (")[0])
    year.append(i1[10].split(" (")[1].split(")")[0])
    score.append(i1[10].split(" (")[1].split(" ")[2].split("%")[0])
    director.append(i1[-1].split(": ")[1])
    cast.append(i1[-3].split(": ")[1])
    critics.append(i1[-6].split(": ")[-1])
```

```python
netflix_movies = {"Rank": rank,"Movie Name":title,"Release Year":year,"Movie Rating":score,"Director":director,"Cast":cast,"Critics Consensus":critics}
```

```python
Netflix_Movies = pd.DataFrame(netflix_movies)
Netflix_Movies.head()
```

```python
Netflix_Movies.to_csv("Top 100 Netflix Movies.csv",index = False)
```
